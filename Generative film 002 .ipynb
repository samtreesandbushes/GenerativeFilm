```python
########################################################################################
# Automate process across an entire directory
########################################################################################

########################################################################################


directory = "eiderfilms"

allClips = []

# iterate over directory
for filename in os.listdir(directory):
    
    print(filename)
    
    if filename.lower().endswith(('.mp4', '.mkv', '.avi')):

        # create array of videoclips
        temp = VideoFileClip(directory+"/"+filename) 
        allClips.append(temp)
        
        

        
final_clip = concatenate_videoclips(allClips, method="compose")
final_clip.write_videofile("all2.mp4",
                           audio_bitrate="1000k", 
                           bitrate="4000k",
                           codec='libx264', 
                           audio_codec='aac', 
                           )

```
